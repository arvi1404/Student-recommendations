```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 

def dropNull(df):
  null_cols = df.columns[df.isnull().all()]
  print("No. of Null Columns : ", len(null_cols))
  if len(null_cols) > 0:
    print("Dropped Columns : ",list(null_cols))
    df.drop(columns=null_cols, inplace=True)
  else : print("No columns were dropped")

def dropDupes(df):
    dupes = df.duplicated().sum()
    print("No. of Duplicate Rows: ", dupes)
    if dupes > 0:
        print("Dropping Duplicates Rows...")
        df.drop_duplicates(inplace=True)
        df = df.reset_index(drop=True)
    else : print("No rows were dropped")

def preprocess_data(df_original):
    df = df_original.copy()
    # Convert date columns to datetime
    date_cols = ['submitted_at', 'created_at', 'updated_at', 'started_at', 'ended_at', 'quiz_created_at', 'quiz_updated_at', 'quiz_end_time', 'quiz_ends_at', 'quiz_daily_date']
    for col in date_cols:
        if col in df.columns:
            df[col] = pd.to_datetime(df[col], utc=True)
            df[col] = df[col].dt.tz_localize(None)  # Remove timezone info if needed
    
    # Convert accuracy to numeric
    df['accuracy'] = df['accuracy'].astype(str).str.rstrip('%').astype(float) / 100

    df['rank_text'] = df['rank_text'].str.extract(r'(\d+)$')

    df[['minutes', 'seconds']] = df['duration'].str.split(':', expand=True).astype(int)
    df['duration'] = df['minutes'] + df['seconds']/60
    df.drop(columns=['minutes', 'seconds'],inplace=True)
    
    return df
```

```python
historical_df['submitted_at'] = pd.to_datetime(historical_df['submitted_at'])
historical_df = historical_df.sort_values(by='submitted_at')
# Grouping by user_id and aggregating relevant statistics
df_features = historical_df.groupby('user_id').agg(
    total_attempts=('quiz_id', 'count'),
    avg_score=('score', 'mean'),
    avg_accuracy=('accuracy', 'mean'),
    avg_speed=('speed', 'mean'),
    avg_negative_score=('negative_score', 'mean'),
    total_questions_attempted=('total_questions', 'sum'),
    mistakes_corrected_ratio=('mistakes_corrected', lambda x: (x.sum() / (historical_df.loc[x.index, 'initial_mistake_count'].sum() + 1e-5))),
    quiz_variety=('quiz_id', 'nunique'),
    time_between_quizzes=('submitted_at', lambda x: x.diff().dt.total_seconds().div(86400).mean() if len(x) > 1 else 0)
).reset_index()

    
df_features.head()
```

# Insights

```python
import pandas as pd

def analyze_user_weakness(df, user_id):
    # Filter data for the given user
    user_df = df[df["user_id"] == user_id]

    # Convert timestamps if not already done
    user_df["started_at"] = pd.to_datetime(user_df["started_at"])
    user_df["ended_at"] = pd.to_datetime(user_df["ended_at"])

    # Compute time taken in seconds
    user_df["time_taken"] = (user_df["ended_at"] - user_df["started_at"]).dt.total_seconds()

    # Compute time per question
    user_df["time_per_question"] = user_df["time_taken"] / user_df["total_questions"]

    # Compute average time per question per topic
    topic_time_avg = df.groupby("quiz_topic")["time_per_question"].mean().reset_index()
    topic_time_avg.rename(columns={"time_per_question": "topic_time_avg"}, inplace=True)

    # Compute average accuracy per topic
    topic_accuracy_avg = df.groupby("quiz_topic")["accuracy"].mean().reset_index()
    topic_accuracy_avg.rename(columns={"accuracy": "topic_accuracy_avg"}, inplace=True)

    # Compute average score per topic
    user_df["score_percentage"] = user_df["final_score"] / user_df["quiz_total"]
    topic_score_avg = user_df.groupby("quiz_topic")["score_percentage"].mean().reset_index()
    topic_score_avg.rename(columns={"score_percentage": "topic_score_avg"}, inplace=True)

    # Merge user data with topic averages
    user_df = user_df.merge(topic_time_avg, on="quiz_topic", how="left")
    user_df = user_df.merge(topic_accuracy_avg, on="quiz_topic", how="left")
    user_df = user_df.merge(topic_score_avg, on="quiz_topic", how="left")

    # Define weakness conditions
    weak_time = user_df["time_per_question"] > user_df["topic_time_avg"]
    weak_accuracy = user_df["accuracy"] < 0.6  # Weak if accuracy < 60%
    weak_score = user_df["topic_score_avg"] < 0.6

    # Group weak topics by issue type
    weak_time_topics = user_df[weak_time]["quiz_topic"].unique()
    weak_accuracy_topics = user_df[weak_accuracy]["quiz_topic"].unique()
    weak_score_topics = user_df[weak_score]["quiz_topic"].unique()
    
    # Print grouped issues
    print(f"\nðŸ” **Weakness Analysis for User {user_id}**")
    
    if len(weak_time_topics) > 0:
        print("\nâ³ **Time Management Issues:** Work on solving questions faster with timed practice.")
        print("   - Topics:", ", ".join(weak_time_topics))
    
    if len(weak_accuracy_topics) > 0:
        print("\nâŒ **Accuracy Issues:** Review fundamental concepts and attempt more practice questions.")
        print("   - Topics:", ", ".join(weak_accuracy_topics))
    
    if len(weak_score_topics) > 0:
        print("\nðŸ“‰ **Score Issues:** Review mistakes and focus on correcting weak areas.")
        print("   - Topics:", ", ".join(weak_score_topics))
    
    print("\nâœ… **Suggested Approach:** Revise notes, practice questions, and analyze incorrect answers.")


# Example usage
analyze_user_weakness(df_new, user_id="YcDFSO4ZukTJnnFMgRNVwZTE4j42")
```
