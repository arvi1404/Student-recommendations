```python
import json
import pandas as pd

# Load JSON data (if reading from a file)
with open("hist.json", "r") as f:
    quiz_data = json.load(f)
```

```python
quiz_data[0].keys()
```

```python
df = pd.DataFrame(quiz_data)
df.head(2)
```

```python
df['quiz'] = list(df['quiz'])
df.head(2)
```

```python
df_quiz = pd.json_normalize(df['quiz']).add_prefix("quiz_")
df_quiz.head()
```

```python
df_merged = pd.concat([df,df_quiz],axis=1)
df_merged.drop(columns = ['quiz'],inplace=True)
df_merged.head(2)
```

```python
df_merged.info()
```

```python
df_merged.to_csv('clean_data.csv')
```
